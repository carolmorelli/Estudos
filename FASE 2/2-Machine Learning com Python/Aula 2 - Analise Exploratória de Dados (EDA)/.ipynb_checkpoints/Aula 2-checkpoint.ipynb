```python
# Criar nova coluna no dataframe
df['duracao_em_min'] = df['duration_ms'] / 60000
```

```python
# trazer as primeiras 5 linhas
df.head(5)
```

```python
df.describe()
```

```python
# Agrupar os albuns e obter a m√©dia total de m√∫sica.
df.groupby("album")["duracao_em_min"].mean()
```

```python
# video 4
import matplotlib.pyplot as plt
```

```python
# ordernar com maiores m√©dias
df_maior_duracao_musica = df.groupby("album")["duracao_em_min"].mean().sort_values(ascending=False)
```

```python
df_maior_duracao_musica
```

```python
# Pegar o Top 5 e usaar o gr√°fico de barras 
df_maior_duracao_musica.head(5).plot(kind='bar')
plt.xlabel('√Ålbuns') # colocando um t√≠tulo pro eixo horizontal do gr√°fico
plt.ylabel('Dura√ß√£o m√©dia (min)') #colocando um t√≠tulo pro eixo vertical do gr√°fico
plt.title('Top 5 √°lbuns com maior dura√ß√£o m√©dia de m√∫sicas') #T√≠tulo do gr√°fico
plt.show() # mostrar o gr√°fico
```
